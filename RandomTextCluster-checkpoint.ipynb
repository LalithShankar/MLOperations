```python
df1 = pd.read_csv(r'C:\Users\a743661\PycharmProjects\MLOperations\hello.csv', header=0, encoding='iso-8859-1')
df1
```

```python
sentences = df1['q'].tolist()
print(sentences)



     

```

```python
r = []
for i in sentences:
        tokenized_text = word_tokenize(i)
        print(tokenized_text)
        r.append(tokenized_text)

```

```python
print(r)
```

```python

model = Word2Vec(r, min_count=1, size = 100)
print (list(model.wv.vocab))
```

```python
X = model[model.wv.vocab]
```

```python
from nltk.cluster import KMeansClusterer
import nltk
```

```python
NUM_CLUSTERS=20
```

```python
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=25)
```

```python
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
```

```python
print (assigned_clusters)
```

```python
words = list(model.wv.vocab)
x_list=[]
for i, word in enumerate(words):
    d={}
    d['name']=word
    d['value']=assigned_clusters[i]
    x_list.append(d)
    print (word + ":" + str(assigned_clusters[i]))
```
